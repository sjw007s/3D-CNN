```python
from PIL import Image
for i in range(11000):
    source_image = "image_data/"+str(i)+".png"
    resize_image_ = "image_data/"+str(i)+"_resize_80_40_hard.png"
    image = Image.open(source_image)
    # resize 할 이미지 사이즈 
    resize_image = image.resize((80, 40))
    # 저장할 파일 Type : JPEG, PNG 등 
    # 저장할 때 Quality 수준 : 보통 95 사용 
    resize_image.save(resize_image_, "PNG")
```

```python
from PIL import Image
import csv
for i in range(11000):
    im = Image.open("image_data/"+str(i)+".png") #Can be many different formats.
    width, height = im.size
    print(width, height)
    pix = im.load()
    #print (im.size) #Get the width and hight of the image for iterating over
    for j in range(height):
        temp_list_r=list()
        temp_list_g=list()
        temp_list_b=list()
        #print("j",j)
        for k in range(width):
            red, green, blue=pix[k,j]
            temp_list_r.append(red)
            temp_list_g.append(green)
            temp_list_b.append(blue)
            #print("k",k)
        with open(str(i)+"_r_"+str(width)+"_"+str(height)+".csv",'a',newline='') as aa:
            writer = csv.writer(aa, delimiter=',')
            writer.writerow(temp_list_r)
        with open(str(i)+"_g_"+str(width)+"_"+str(height)+".csv",'a',newline='') as aa:
            writer = csv.writer(aa, delimiter=',')
            writer.writerow(temp_list_g)

        with open(str(i)+"_b_"+str(width)+"_"+str(height)+".csv",'a',newline='') as aa:
            writer = csv.writer(aa, delimiter=',')
            writer.writerow(temp_list_b)
                
    
```
